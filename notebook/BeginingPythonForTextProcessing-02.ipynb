#Tokenization

##Counting Word Tokens

### with NLTK

```
import nltk

file_name = "frankenstein-100.txt"
with (open(file_name)) as f:
  content = f.read()
  nltk_tokens = nltk.word_tokenize(content)
  print(nltk_tokens)
  print("\n")
  print("Number of Words: " , len(nltk_tokens))
```

### with Split

```
file_name = "frankenstein-100.txt"
with (open(file_name)) as f:
  content = f.read()
  print(content.split())
  print("\n")
  print("Number of Words: ", len(content.split()))
```

##Uniq Word Tokens

```
file_name = "frankenstein-100.txt"
with (open(file_name)) as f:
  content = f.read()
  nltk_tokens = nltk.word_tokenize(content)
  tokens_set = set(nltk_tokens)
  tokens_list = list(tokens_set)
  tokens_list.sort()
  print(tokens_list)
  print("\n")
  print("Number of Words: " , len(tokens_list))
```

## Counting Uniq Word Tokens

```
import operator

file_name = "frankenstein-100.txt"
token_count = {}
with (open(file_name)) as f:
  content = f.read()
  nltk_tokens = nltk.word_tokenize(content)
  for token in nltk_tokens:
    if(token in token_count):
      token_count[token] += 1
    else:
      token_count[token] = 1

sorted_token_count = sorted(token_count.items(), key=operator.itemgetter(1), reverse = True)
      
for k, v in sorted_token_count:
  print(k + ":" + str(v))
```

##Stopwords & Punctuation Removal

```
nltk.download("stopwords")
```

```
from nltk.corpus import stopwords
stop_words = stopwords.words('indonesian')
print(stop_words)
```

```
from nltk.corpus import stopwords
from string import punctuation

stop_words = stopwords.words('english')

token_filtered = []

file_name = "frankenstein-100.txt"
with (open(file_name)) as f:
  content = f.read()
  nltk_tokens = nltk.word_tokenize(content)
  for token in nltk_tokens:
    if token not in stop_words:
      if token not in punctuation:
        token_filtered.append(token)
        
print(token_filtered)
print(len(token_filtered))
```

#Morphological Analysis

## POS Tag

```
from nltk.corpus import state_union
from nltk.tokenize import PunktSentenceTokenizer
file_name = "frankenstein-100.txt"
data = []
with (open(file_name)) as f:
  content = f.read()
  sentences = nltk.sent_tokenize(content)
  for sent in sentences:
    data = data + nltk.pos_tag(nltk.word_tokenize(sent))

for word in data:
  print(word)
```

##NE Tag

```
file_name = "frankenstein-100.txt"
data = []
with (open(file_name)) as f:
  content = f.read()
  sentences = nltk.sent_tokenize(content)
  for sent in sentences:
    data = data + nltk.pos_tag(nltk.word_tokenize(sent))

  ne_data = nltk.ne_chunk(data)

for word in ne_data:
  print(word)
```

#TF/IDF

```
from nltk.text import TextCollection

text1 = ''
text2 = ''

with (open('hamlet.txt')) as f1:
  text1 = f1.read()
with (open('frankenstein.txt')) as f2:
  text2 = f2.read()
  
texts = TextCollection([text1, text2])

print(texts.tf('HAMLET', text1))
print(texts.idf('HAMLET'))
print(texts.tf_idf('HAMLET', text1))
```
